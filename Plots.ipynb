```python
import pandas as pd
import matplotlib.pyplot as mp
import numpy as np

with open('histogramm.txt') as f:
        lines = f.readlines()
        x=[line.split()[0] for line in lines]
        
```

```python
def plot(N):
    x_t = x[N].split(',')
    x_t = x_t[:-1]
    x_t = list(map(float, x_t))
    x_z = np.array(x_t)
    y = [i for i in range(len(x_t))]
    y_t = np.array(y)
    mp.plot(y_t,x_z)
```
